This notebook allows you to receive links to images from the dataset.

```python
def convert_to_url(signature):
    prefix = 'http://i.pinimg.com/400x/%s/%s/%s/%s.jpg'
    return prefix % (signature[0:2], signature[2:4], signature[4:6], signature)

entry = {
    "product": "0027e30879ce3d87f82f699f148bff7e", 
    "scene": "e7d32df9f45b691afc580808750f73ca", 
    "bbox": [
        0.434097, 
        0.859363, 
        0.560254, 
        1.0
    ]
}

product_url = convert_to_url(entry["product"])
scene_url = convert_to_url(entry["scene"])

print("Product URL:", product_url)
print("Scene URL:", scene_url)
```
